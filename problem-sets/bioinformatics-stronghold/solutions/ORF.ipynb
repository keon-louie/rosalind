```python
with open("../problem_files/rosalind_orf.txt", 'r') as f:
    dna = {}
    seq_names = []
    seqs = []
    buf = ''
    for line, content in enumerate(f):
        if line == 0:
            seq_names.append(content[1:].rstrip())
        elif content.startswith('>') and line > 0:
            seq_names.append(content[1:].rstrip())
            seqs.append(buf)
            buf = ''
        else:
            buf += content.rstrip()
    seqs.append(buf)
    for m, n in zip(seq_names, seqs):
        dna[m] = n

    watson_rna = seqs[0].replace('T', 'U')
    complement_table = watson_rna.maketrans('AUCG', 'UAGC')
    crick_rna = watson_rna.translate(complement_table)[::-1]

print(watson_rna)
print(crick_rna)
```

```python
# ORF dict for Watson
# 0, 1, 2 = ORF shift (start at 0, 1, or 2nd base).
# The values correspond to each of the indices of the start codons within that ORF shift.
orf_frames = {0: [], 1: [], 2: []}
for m in re.finditer('AUG', watson_rna):
    frame = m.start() % 3
    orf_frames[int(frame)].append(m.start())

orfs = [] # Create ORFS list

# Watson 5' to 3'
for i in range(3): # For 3 RFs
    for j in range(len(orf_frames[i])): # Number of start codons within RF
        candidate_orf = ''
        start_here = orf_frames[i][j]
        for k in range(start_here, len(watson_rna) - 2 - i, 3): # iterate every codon (3 bp)
            codon = watson_rna[k:k+3]
            if aa[codon] == 'Stop':
                orfs.append(candidate_orf)
                break
            candidate_orf += aa[codon]

# ORF dict for Crick
orf_frames = {0: [], 1: [], 2: []}
for m in re.finditer('AUG', crick_rna):
    frame = m.start() % 3
    orf_frames[int(frame)].append(m.start())

for i in range(3): # For 3 RFs
    for j in range(len(orf_frames[i])): # Number of start codons within RF
        candidate_orf = ''
        start_here = orf_frames[i][j]
        for k in range(start_here, len(crick_rna) - 2 - i, 3): # iterate every codon (3 bp)
            codon = crick_rna[k:k+3]
            if aa[codon] == 'Stop':
                orfs.append(candidate_orf)
                break
            candidate_orf += aa[codon]

for seq in list(set(orfs)):
    print(seq)
```
