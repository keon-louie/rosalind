```python
import math

k = 17
m = 24
n = 19
t = k + m + n

kk = math.comb(k, 2)
mm = math.comb(m, 2)
nn = math.comb(n, 2)
km = math.comb(k, 1) * math.comb(m, 1)
kn = math.comb(k, 1) * math.comb(n, 1)
mn = math.comb(m, 1) * math.comb(n, 1)

prob_dom = (kk * 1 + mm * 3/4 + nn * 0 + km * 1 + kn * 1 + mn * 1/2) / math.comb(t, 2)
print(prob_dom)
```
