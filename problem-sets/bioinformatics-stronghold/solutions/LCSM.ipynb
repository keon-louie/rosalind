```python
import time
```

```python
with open("../problem_files/rosalind_lcsm.txt", 'r') as f:
    dna = {}
    seq_names = []
    seqs = []
    buf = ''
    for line, content in enumerate(f):
        if line == 0:
            seq_names.append(content[1:].rstrip())
        elif content.startswith('>') and line > 0:
            seq_names.append(content[1:].rstrip())
            seqs.append(buf)
            buf = ''
        else:
            buf += content.rstrip()
    seqs.append(buf)
    for m, n in zip(seq_names, seqs):
        dna[m] = n
```

```python
# This is the SLOW method of finding the biggest shared motif because we will iterate over every possible motif in the parent.

def shared_motif_small_to_big(seqs_list):
    parent_seq = seqs_list[0]
    seqs_list = seqs_list[1:]
    biggest_shared_motif = ''
    t0 = time.time()
    for i in range(1, len(parent_seq) + 1):
        motif_size = i
        if motif_size % 50 == 0:
            print(motif_size)
        for j in range(len(parent_seq) - motif_size + 1):
            motif = parent_seq[j:j+motif_size]
            motif_count = 0
            for seq in seqs_list:
                if motif in seq:
                    motif_count += 1
            if motif_count == len(seqs_list) and len(motif) >= len(biggest_shared_motif):
                biggest_shared_motif = motif
    t1 = time.time()
    return biggest_shared_motif, t1-t0

s_to_b = shared_motif_small_to_big(seqs)
print(s_to_b[0])
print(s_to_b[1])
```

```python
# This is the FASTER method of finding the biggest shared motif because we are searching from the biggest shared motif first, meaning that we stop once we find any motif.

def shared_motif_big_to_small(seqs_list):
    parent_seq = seqs_list[0]
    seqs_list = seqs_list[1:]
    t0 = time.time()
    for i in range(len(parent_seq)):
        motif_size = len(parent_seq) - i
        if motif_size % 50 == 0:
            print(motif_size)
        for j in range(len(parent_seq) - motif_size + 1):
            motif = parent_seq[j:j+motif_size]
            motif_count = 0
            for seq in seqs_list:
                if motif in seq:
                    motif_count += 1
            if motif_count == len(seqs_list):
                t1 = time.time()
                return motif, t1-t0

b_to_s = shared_motif_big_to_small(seqs)
print(b_to_s[0])
print(b_to_s[1])
```
