```python
with open("problem_files/rosalind_gc.txt", 'r') as fasta:
    lines = fasta.readlines()

labels = []
bp = []
gc_contents = []
string = ""

for index, line in enumerate(lines):
    line = line.replace('\n', '')

    if line.startswith('>'):
        line = line.replace('>', '')
        labels.append(line)
        if index != 0:
            bp.append(string)
        string = ''
    elif (index != len(lines) - 1):
        string += line
    else:
        string += line
        bp.append(string)

# print(labels)
# print(bp)

for seq in bp:
    gc_count = 0
    total_count = 0
    gc_content = 0
    for base in seq:
        if base == 'C' or base == 'G':
            gc_count += 1
        total_count += 1
    gc_content = gc_count / total_count * 100
    gc_contents.append(gc_content)

# print(gc_contents)

max_content = gc_contents[0]
max_index = 0
for index, content in enumerate(gc_contents):
    if content > max_content:
        max_content = content
        max_index = index

print(labels[max_index] + '\n' + str(max_content))
```
