```python
with open("../problem_files/rosalind_sseq.txt", 'r') as f:
    dna = {}
    seq_names = []
    seqs = []
    buf = ''
    for line, content in enumerate(f):
        if line == 0:
            seq_names.append(content[1:].rstrip())
        elif content.startswith('>') and line > 0:
            seq_names.append(content[1:].rstrip())
            seqs.append(buf)
            buf = ''
        else:
            buf += content.rstrip()
    seqs.append(buf)
    for m, n in zip(seq_names, seqs):
        dna[m] = n
```

```python
parent = seqs[0]
motif = seqs[1]
indices = ''
index = 0
index_length = 0

for base in motif:
    index_length += index
    index = parent[index_length:].find(base) + 1
    print(index + index_length, end = ' ')
```
