```python
with open("../problem_files/rosalind_subs.txt", 'r') as f:
    sample = f.readline().rstrip()
    substring = f.readline().rstrip()

repeats = ''

for i in range(0, len(sample) - len(substring) + 1):
    string_compare = sample[i:i+len(substring)]
    if string_compare == substring:

        repeats += str(i + 1) + ' '

print(repeats)
```
