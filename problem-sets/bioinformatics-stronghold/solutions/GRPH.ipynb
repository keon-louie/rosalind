```python
with open("../problem_files/rosalind_grph.txt", 'r') as f:
    dna = {}
    seq_names = []
    seqs = []
    buf = ''
    for line, content in enumerate(f):

        if content.startswith('>') and line > 0:
            seq_names.append(content[1:].rstrip())
            seqs.append(buf)
            buf = ''
        elif line == 0:
            seq_names.append(content[1:].rstrip())
        else:
            buf += content.rstrip()
    seqs.append(buf)
    for i, j in zip(seq_names, seqs):
        dna[i] = j
```

```python
seq_names, seqs = dna.keys(), dna.values()
adj_list = [[]]

n1 = list(seqs)
n2 = list(seqs)
seq_names = list(seq_names)

for i1, seq1 in enumerate(n1):
    for i2, seq2 in enumerate(n2):
        if seq1[-3:] == seq2[:3] and seq1 != seq2:
            print(seq_names[i1], seq_names[i2])
```
