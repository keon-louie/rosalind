```python
import numpy as np
```

```python
# Reading FASTA file
with open("../problem_files/rosalind_cons.txt", 'r') as f:
    seq = []
    buf = ''
    labels = []
    for line, content in enumerate(f):
        if line == 0:
            labels.append(content[1:].rstrip())
        elif content.startswith('>') and line > 0:
            labels.append(content[1:].rstrip())
            seq.append(buf)
            buf = ''
        else:
            buf += content.rstrip()
    seq.append(buf)

size = len(seq[0])
profile = {'A': [0] * size,
           'T': [0] * size,
           'C': [0] * size,
           'G': [0] * size}
profile_matrix = np.zeros((4, size))
consensus = ''

for string in seq:
    for index, base in enumerate(string):
        profile[base][index] += 1

# Creating profile matrix
for string in seq:
    for string_index, base in enumerate(string):
        if base == 'A':
            profile_matrix[0, string_index] += 1
        elif base == 'C':
            profile_matrix[1, string_index] += 1
        elif base == 'G':
            profile_matrix[2, string_index] += 1
        elif base == 'T':
            profile_matrix[3, string_index] += 1
        else:
            print("Base not found in ATCG")

profile_matrix_transpose = profile_matrix.T

# Creating consensus
for row in profile_matrix_transpose:
    consensus += str(np.argmax(row))

# Turn consensus into bp's...
consensus = consensus.replace('0', 'A')
consensus = consensus.replace('1', 'C')
consensus = consensus.replace('2', 'G')
consensus = consensus.replace('3', 'T')

# Print results
print(consensus)
bases = 'ACGT'
for index, row in enumerate(profile_matrix):
    print("{}: ".format(bases[index]), end = "")
    for item in row:
        print(str(int(item)), end = " ")
    print("")
```
