```python
N = 80739
gc = 0.468213
at = 1 - gc
motif = 'TTAGGATAC'
no_motif_sum = 0

# Find motif chance for random sequence of length motif
motif_chance = 1
for base in motif:
    if base == 'A' or base == 'T':
        motif_chance *= at / 2
    else:
        motif_chance *= gc / 2

no_motif_sum = (1 - motif_chance) ** (N - len(motif) + 1)
one_plus_motif = 1 - no_motif_sum
print(one_plus_motif)
```
