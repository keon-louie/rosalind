```python
# tldr: numpy is not good for large arithmetic, so this matrix using numpy array won't work...

def rabbit_matrix(n, m):
# number of months that each rabbit lives # number of months until we measure population

    rabbits = np.zeros((m, n))
    rabbits[0:1, 0:1] = 1
    rabbits[1:2, 1:2] = 1

    # print("Original rabbit matrix: \n{}".format(str(rabbits) + '\n'))

    for col in range(2, rabbits.shape[1]):
        r_adult_count = 0
        for row in range(1, rabbits.shape[0]):
            r_adult_count += rabbits[row, col - 1]
            rabbits[row, col] = rabbits[row - 1, col - 1]
        rabbits[0, col] = r_adult_count
    # print("Number of rabbits at month n: {}".format(int(np.sum(rabbits[:, -1]))))
    return rabbits
    # print("New rabbit matrix: \n{}\n".format(rabbits))
```

```python
n, m = 100, 17
fib = [0] * n # INSTEAD OF np.zeros(n) BECAUSE OF THE STUPID ROUNDING ERROR
fib[0] = 1
fib[1] = 1

# for the first m rabbit population, it will follow the fibonacci sequence (as there are no dead rabbits). thus,
for k in range(2, m):
    fib[k] = fib[k-1] + fib[k-2]

# print(fib)

# for the m+1 to n population
for j in range(m, n):
    for k in range(j - m, j - 1):
        fib[j] += fib[k]

print(int(fib[-1]))
```
