```python
def create_elements(n):
    p = []
    for k in range(n):
        p.append(n - k)
    return p

def gen_perm_manual(elements):

    if len(elements) == 0:
        return [[]]

    elif len(elements) == 1:
        return [[elements[0]]]

    all_permutations = []

    for i in range(len(elements)):
        fixed = elements[i]
        remaining = elements[:i] + elements[i+1:]
        perm_remaining = gen_perm_manual(remaining)

        for perm in perm_remaining:
            all_permutations.append([fixed] + perm)

    return all_permutations

n = 6
factorial = 1
p = create_elements(n)
permutations = gen_perm_manual(p)

for num in range(n):
    factorial *= (n - num)

print(factorial)
for perm in permutations:
    for item in perm:
        print("{} ".format(item), end = "")
    print("")
```

```python
def generate_permutations_manual(elements):
    """
    Generates all permutations of a list of elements manually using recursion.

    Args:
        elements: A list of elements for which to generate permutations.

    Returns:
        A list of lists, where each inner list represents a permutation.
    """
    if len(elements) == 0:
        return [[]]  # Base case: empty list has one permutation (empty list)

    if len(elements) == 1:
        return [[elements[0]]] # Base case: single element list has one permutation (itself)

    all_permutations = []
    for i in range(len(elements)):
        # Take the current element as the "fixed" element
        fixed_element = elements[i]

        # Create a list of remaining elements
        remaining_elements = elements[:i] + elements[i+1:]

        # Recursively get permutations of the remaining elements
        permutations_of_remaining = generate_permutations_manual(remaining_elements)

        # Combine the fixed element with each permutation of the remaining elements
        for perm in permutations_of_remaining:
            all_permutations.append([fixed_element] + perm)

    return all_permutations

# Example usage:
my_list = [1, 2, 3]
permutations = generate_permutations_manual(my_list)
print(f"Permutations of {my_list}:")
for p in permutations:
    print(p)
```
