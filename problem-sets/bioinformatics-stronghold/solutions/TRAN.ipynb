```python
with open("../problem_files/rosalind_tran.txt", 'r') as f:
    seq = []
    buf = ''
    labels = []
    for line, content in enumerate(f):
        if line == 0:
            labels.append(content[1:].rstrip())
        elif content.startswith('>') and line > 0:
            labels.append(content[1:].rstrip())
            seq.append(buf)
            buf = ''
        else:
            buf += content.rstrip()
    seq.append(buf)

a = seq[0]
b = seq[1]
purine = 'AG'
pyrimidine = 'CT'

hamming = 0
transition = 0
transversion = 0
for base in range(len(a)):
    if a[base] != b[base]:
        if a[base] in purine and b[base] in purine:
            transition += 1
        elif a[base] in pyrimidine and b[base] in pyrimidine:
            transition += 1
        elif a[base] in purine and b[base] in pyrimidine:
            transversion += 1
        elif a[base] in pyrimidine and b[base] in purine:
            transversion += 1

ratio = transition / transversion
print(ratio)
```
