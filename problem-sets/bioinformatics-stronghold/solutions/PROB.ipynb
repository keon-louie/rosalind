```python
import math
```

```python
with open("../problem_files/rosalind_prob.txt", 'r') as f:
    target = f.readline().rstrip()
    gc = f.readline().rstrip()
```

```python
gc = gc.split(' ')
B = []
for i in range(len(gc)):
    prob = 1
    gc[i] = float(gc[i])
    for base in target:
        if base == 'A':
            prob *= (1 - gc[i]) / 2
        elif base == 'C':
            prob *= gc[i] / 2
        elif base == 'G':
            prob *= gc[i] / 2
        else:
            prob *= (1 - gc[i]) / 2
    print(round(math.log10(prob), 3), end = ' ')
```
