```python
with open("../problem_files/rosalind_splc.txt", 'r') as f:
    seq = []
    buf = ''
    labels = []
    for line, content in enumerate(f):
        if line == 0:
            labels.append(content[1:].rstrip())
        elif content.startswith('>') and line > 0:
            labels.append(content[1:].rstrip())
            seq.append(buf)
            buf = ''
        else:
            buf += content.rstrip()
    seq.append(buf)

print(labels)
print(seq)
```

```python
p_mrna = seq[0]
mrna = ''
introns = seq[1:]

for intron in introns:
    while p_mrna.find(intron) != -1:
        # print("we found an intron at index {} for {}".format(p_mrna.find(intron), intron))
        index = p_mrna.find(intron)
        p_mrna = p_mrna[:index] + p_mrna[index+len(intron):]

print(p_mrna)
```
