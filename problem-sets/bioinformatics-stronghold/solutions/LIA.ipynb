```python
import math
import scipy as scp
import numpy as np
```

```python
chance_scp = 0
chance_manual = 0
n = 31
k = 7
num_children = 2 ** k
for i in range(n, num_children + 1):
    chance_scp += scp.stats.binom.pmf(i, num_children, 0.25)
    chance_manual += (0.25 ** i) * (0.75 ** (num_children - i)) * (math.comb(num_children, i))

print(chance_manual)
print(chance_scp)
```
